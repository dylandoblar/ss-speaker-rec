### Functions for working with concatenated consecutive time and audio chunks

```python
def create_ann(valid_segs, labels, audacity_file=None):
    ann = pyannote.core.Annotation()

    if len(valid_segs) != len(labels):
        print("Error: Number of segments to be labeled does not match number of labels.")
        exit()

    seg_start, seg_end = valid_segs[0]
    for i in range(len(valid_segs)-1):
        next_seg_start, next_seg_end = valid_segs[i+1]
        if (seg_end == next_seg_start) and (labels[i] == labels[i+1]):
            seg_end = next_seg_end
        else:
            ann_seg = pyannote.core.Segment(seg_start, seg_end)
            ann[ann_seg] = labels[i]
            seg_start, seg_end = valid_segs[i+1]
    
    ann_seg = pyannote.core.Segment(seg_start, seg_end)
    ann[ann_seg] = labels[i]

    return ann

def export_ann(ann, save_file=None):
    if save_file != None:
        filename = save_file
    else:
        filename = 'annotation.txt'
        
    with open(filename, 'w') as f:
        for seg in ann.itersegments():
            f.write(str(seg[0])+'\t'+str(seg[1])+'\t'+str(ann[seg])+'\n')
    f.close()
    return
```

### Process audio and pass through speech embedder model

```python
speaker1 = '../data/sample_timit/speaker1.wav'
speaker2 = '../data/sample_timit/speaker2.wav'
speaker3 = '../data/sample_timit/speaker3.wav'
speaker4 = '../data/sample_timit/speaker4.wav'

model_path = './timit.model'

# Set up speech embedder model
embedder_net = SpeechEmbedder()
embedder_net.load_state_dict(torch.load(model_path))
embedder_net.eval()
print("Model loaded.")
```

```python
def get_concat_segs

# Detect voice activity in the audio file.
times, segs = VAD_chunk(2, audio_path)
if segs == []:
    print('No voice activity detected in audio file!')
    exit()

# Concatenate consecutive frames of audio and time
print("Concatenating consecutive audio and time frames.")
concat_seg, concat_times = concat_times_and_chunks(times, segs)

# Get total duration of speech in audio file
duration = round(get_speech_duration(concat_times), 2)
print("Total speech duration is %f seconds." % duration)
```

```python
# Compute STFTs of voiced audio segments. (240ms window, 120ms overlap)
print("Computing STFTs.")
STFTs_by_chunk = get_STFTs_by_chunk(concat_seg)
```

```python
STFTs_by_chunk[0].shape
```

```python
avg_embeddings = []
for seg in STFTs_by_chunk:
    emb = embedder_net(seg)
    seg_averaged = avg_dvectors(emb.detach().numpy())
    avg_embeddings.append(seg_averaged)
    del emb
stacked = np.vstack(avg_embeddings)
```

```python
stacked.shape
```

```python
# Get segments with frames at end that are too short for embeddings and
# get timestamps only for valid segments.
mismatches = get_mismatched_frames(concat_times, avg_embeddings)
valid_segs = get_valid_segs(concat_times, mismatches)
```

```python
# Nifty piece of code for debugging mismatched frame and label counts

missing_idx = set()
label_count = 0
expected = {}
for i, seg in enumerate(concat_times):
    duration = round(seg[1],2)-round(seg[0],2)
    num_frames = math.ceil(round(duration/0.4,2))
    label_count += num_frames
    expected[seg] = num_frames
    if num_frames != avg_embeddings[i].shape[0]:
        missing_idx.add(i)
        if avg_embeddings[i].shape[0] - num_frames != -1:
            print(i, seg, num_frames, avg_embeddings[i].shape[0])

# print(label_count)
# print(len(missing_idx))
```

### 2-dimensional t-SNE visualization of embeddings

```python
tsne_emb_2 = sklearn.manifold.TSNE(
    n_components=2,
    random_state=42
).fit_transform(stacked)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(tsne_emb_2[:,0], tsne_emb_2[:,1])
plt.show()
```

### 2-dimensional UMAP visualization of embeddings

```python
import umap

n_components = 50
umap_emb_n = umap.UMAP(
    metric='cosine',
    n_components=n_components,
    random_state=42,
).fit_transform(stacked)

umap_emb_2 = umap.UMAP(
    metric='cosine',
    n_components=2,
    random_state=42,
).fit_transform(stacked)
```

```python
plt.figure()
plt.scatter(umap_emb_2[:,0], umap_emb_2[:,1])
plt.show()
```
